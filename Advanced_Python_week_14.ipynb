# 03 ANN improve 

```python
def house_model(y_new):
    xs = np.array([0, 1, 2, 3, 4, 5, 6], dtype=float)
    ys = np.array([50, 100, 150, 200, 250, 300, 350], dtype=float)
    
    model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
    
    model.compile(optimizer='sgd', loss='mean_squared_error')
    
    model.fit(xs, ys, epochs=500)
    
    return model.predict(y_new)
```

```python
house_model([17])
```

```python
def house_model_scaled(y_new):
    xs = np.array([0, 1, 2, 3, 4, 5, 6], dtype=float)
    ys = np.array([0.5, 1, 1.5, 2, 2.5, 3, 3.5], dtype=float)
    
    model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
    
    model.compile(optimizer='sgd', loss='mean_squared_error')
    
    model.fit(xs, ys, epochs=500)
    
    return 100 * model.predict(y_new)[0][0]
```

```python
import matplotlib.pyplot as plt
```

# 4-1 load data

```python
fashion_mnist = keras.datasets.fashion_mnist
```

```python
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

```python
# class names define
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

# 4-3 preprocessing

```python
train_images = train_images / 255.0
test_images = test_images / 255.0
```

```python
plt.figure()
plt.imshow(train_images[0])
plt.colorbar()
plt.show()
```

# 4-4 model train

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])
```

```python
model.compile(optimizer = 'adam',
             loss = 'sparse_categorical_crossentropy',
             metrics=['accuracy'])
```

```python
model.fit(train_images, train_labels, epochs=50)
```

# 4-5 Model performance

```python
test_loss, test_acc = model.evaluate(test_images, test_labels)
```

```python
print('test accuracy: ', test_acc)
```
