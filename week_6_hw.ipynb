```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('train.csv')
```

```python
data.info()
```

```python
data.head(20)
```

```python
## 1-1 encoding
#MSZoning column을 정수로 인코딩 해봅시다.
```

```python
MSZoning_mapping = {'RL':0, 'RM':1}
data['MSZoning'] = data['MSZoning'].map(MSZoning_mapping)
```

```python
data.head(20)
```

```python
#visulization
```

```python
plt.figure(figsize=(4,10))
sns.boxplot(data['MSZoning'])
```

```python
#Outlier remove
```

```python
print('skewness of MSZoning:', data['MSZoning'].skew())# It is outlier
```

```python
Q1 = data['MSZoning'].quantile(0.25)
Q3 = data['MSZoning'].quantile(0.75)
IQR = Q3 - Q1

outliers = data[(data['MSZoning'] < (Q1 - 1.5 * IQR)) | (data['MSZoning'] > (Q3 + 1.5 * IQR))]

```
