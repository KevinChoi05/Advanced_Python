# 00 Import module, load data

```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = load_iris()

features = iris['data']
feature_names = iris['feature_names']
label = iris['target']

df = pd.DataFrame(features, columns=feature_names)
df['target'] = label
```

# 01 Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

# 02 K fold cross validation (CV)

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
kfold = KFold(n_splits=5)
```

```python
features
```

```python
label
```

```python
scores = cross_val_score(rf, features, label, cv=kfold)
```

```python
print('fold accuracy: ', scores)
```

```python
print('fold mean accuracy: ', np.mean(scores))
```

# 03 Frid search & K fold CV

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'n_estimators' : [100, 150, 200, 250],
    'max_depth' : [None, 6, 9, 12],
    'min_samples_split' : [0.01, 0.05, 0.1],
    'max_features' : ['auto', 'sqrt']
}
```

# 3-1 Create Model

```python
rf2 = RandomForestClassifier()
```

# 3-2 Create K fold

```python
kfold2 = KFold(n_splits = 10)
```

```python
grid_search = GridSearchCV(estimator = rf2,
                           param_grid = param_grid,
                           cv = kfold2)
```

```python
grid_search.fit(features, label)
```

```python
grid_search.best_params_
```
