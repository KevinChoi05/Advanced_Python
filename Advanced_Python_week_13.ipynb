# 02 Activation function

```python
#sigmoid
```

```python
def sigmoid(x):
    return 1 / (1+np.exp(-x))
```

```python
x = np.arange(-5, 5, 0.1)
```

```python
x
```

```python
y = sigmoid(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
#step function
```

```python
def step_function(x):
    return np.array(x>0)
```

```python
y = step_function(x)

plt.plot(x,y)
plt.show()
```

```python
#relu function
```

```python
def relu(x):
    return np.maximum(0, x)
```

```python
y = relu(x)

plt.plot(x,y)
plt.show()
```

```python
y = relu(x)
plt.plot(x,y, label='relu')
y = step_function(x)
plt.plot(x,y,label='step')
y = sigmoid(x)
plt.plot(x,y, label='sigmoid')
plt.legend()
plt.show()
```
